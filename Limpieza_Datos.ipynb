```python
#Codigo para crear una lista del diccionario que nos dan
import pandas as pd
excel1='C:/BD_Diccionario.csv'
df2 = pd.read_csv(excel1,sep =";")
Lista_Nombres = df2['Nombre'].tolist()
#Lista_Nombres
#Codigo para comparar los encabezados de la tabla con la lista del diccionario
excel2='C:/BD_Limpieza_Ejercicio.csv'
df = pd.read_csv(excel2,sep =";")
importar_encabezados = df.axes[1]
#print(importar_encabezados)
for i in importar_encabezados:
    if i not in Lista_Nombres:
        df = df.drop(columns=i)
        #print(i)
#Codigo que nos indica cuanta informacion Null tiene cada columna
#df.replace(["999"],np.nan,inplace=True)
#df.replace(["."],np.nan,inplace=True)
for column in df.columns:
    #df[column]=df[column].astype(str).str.upper()
    df[column]=df[column].astype(str).str.replace("à","O")
    df[column]=df[column].astype(str).str.replace("µ","Á")
    df[column]=df[column].astype(str).str.replace("NN","N")
    df[column]=df[column].astype(str).str.replace("@","A")
    df[column]=df[column].astype(str).str.replace("Ö","Í")
    df[column]=df[column].astype(str).str.replace("CC","C")
    df[column]=df[column].astype(str).str.replace("¥","NI")
    df[column]=df[column].astype(str).str.replace("II","I")
    df[column]=df[column].astype(str).str.replace("COTA","COSTA")
    df[column]=df[column].astype(str).str.replace("TULCMN","TULCAN")
    df[column]=df[column].astype(str).str.replace("D3LEG","DÉLEG")
    df[column]=df[column].astype(str).str.replace("CONCLUIDO","TERMINADO")

#for column in df.columns:
 #   Lista = df[column].tolist()
  #  Set = set(Lista)
df.dropna()
df
```
